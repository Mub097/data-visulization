```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,10)
```

```python
y=x*2
```

```python
#labeling the plot with names
```

```python
plt.plot(x,y,color='b',linestyle=':',linewidth=2)
plt.title(" x vs y ")
plt.xlabel("this xlabel")
plt.ylabel("this y label")
plt.show()
```

```python
#how to add two lines in one plot
```

```python
x = np.arange(1,11)
```

```python
y1=x*2
```

```python
y2=x*3
```

```python
plt.plot(x,y1,color='g',linestyle=":")
plt.title("garph")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.grid(True)
plt.plot(x,y2,color="y")
plt.show()
```

```python
#creating subplots
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='g',linestyle=":")
plt.title("graph")
plt.xlabel("this is x axis")
plt.subplot(2,1,2)
plt.ylabel("this is y axis")
plt.grid(True)
plt.plot(x,y2,color="y")
plt.show()
```
